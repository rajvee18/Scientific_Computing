```python
import numpy as np
n=int(input("enter the size of the matrix"))


A = np.zeros((n,n), dtype=float)
l=np.eye(n,n)
u=np.zeros(n,n)
for i in range(n):
    for j in range(n):
        A[i,j] = float(input("enter the element rowwise:"))

for i in range(n):
    for j in range(i,n):
    
        u[i][j]=A[i][j]-np.dot(l[i,:i],u[:i,j])

    for j in range(i+1,n):
        l[i+1][j] = (A[i][j] - np.dot())/A[i][i]

        

```
